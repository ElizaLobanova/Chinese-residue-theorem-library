```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('time sequential18.csv', 'r') as f:
    lines = f.readlines()
    
    gmp_time = []
    crt_time = []
    data_size = []
    lunch_times = []
    for line in lines:
        gmp_time.append(float(line.split(' ')[0]))
        crt_time.append(float(line.split(' ')[1]))
        data_size.append(float(line.split(' ')[2]))
        lunch_times.append(float(line.split(' ')[3]))
    gmp_time = np.array(gmp_time)
    crt_time = np.array(crt_time)
    data_size = np.array(data_size)
    lunch_times = np.array(lunch_times)
    print(gmp_time, crt_time, data_size, lunch_times)
    add_p_ms_gmp = (data_size * lunch_times) / (gmp_time / 1000)
    add_p_ms_crt = (data_size * lunch_times) / (crt_time / 1000)
    plt.xlabel('длина массива', fontsize=18)
    plt.ylabel('операции / сек', fontsize=18)
    plt.scatter(data_size, add_p_ms_gmp, color='red', label='GMP', marker='+')
    plt.scatter(data_size, add_p_ms_crt, color='green', label='КТО')
    plt.legend(loc='lower right')
    plt.xscale('log')
    plt.xticks(size=14)
    plt.yticks(size=14)
    plt.yscale('log')
    plt.savefig('../../time_sequential18.pdf')
    plt.show()
```
